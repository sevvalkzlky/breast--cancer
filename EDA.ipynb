```python
# IQR yöntemiyle aykırı değerleri tespit edelim
Q1 = df['radius_mean'].quantile(0.25)
Q3 = df['radius_mean'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Aykırı değerleri dışarıda bırak
df = df[(df['radius_mean'] >= lower_bound) & (df['radius_mean'] <= upper_bound)]
```

```python
# Aykırı değerleri medyanla değiştir
median = df['radius_mean'].median()

df['radius_mean'] = df['radius_mean'].apply(
    lambda x: median if x < lower_bound or x > upper_bound else x
)
```

```python
df['radius_mean'] = df['radius_mean'].apply(
    lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x
)
```

```python
# Korelasyonu yüksek olan çiftleri bul
threshold = 0.9  # çok yüksek korelasyon sınırı
high_corr_pairs = []

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            colname1 = corr_matrix.columns[i]
            colname2 = corr_matrix.columns[j]
            corr_value = corr_matrix.iloc[i, j]
            high_corr_pairs.append((colname1, colname2, corr_value))

# Sonuçları yazdır
print("Yüksek korelasyonlu özellik çiftleri (>|0.9|):")
for col1, col2, corr_val in high_corr_pairs:
    print(f"{col1} - {col2}: {corr_val:.2f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Veri ve hedef
X = df.drop(['id', 'diagnosis'], axis=1)  # Özellikler
y = df['diagnosis']  # Hedef (benign/malignant)

# Eğer diagnosis kategorik ise label encoding yapalım
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_enc = le.fit_transform(y)  # B=0, M=1 gibi sayısal değerler

# Veri bölme
X_train, X_test, y_train, y_test = train_test_split(
    X, y_enc, test_size=0.2, random_state=42, stratify=y_enc)

# Model oluşturma
rf = RandomForestClassifier(
    n_estimators=100,
    class_weight='balanced',  # Dengesiz sınıflar için ağırlıklandırma
    random_state=42)

# Modeli eğit
rf.fit(X_train, y_train)

# Tahmin yap
y_pred = rf.predict(X_test)

# Sonuçları yazdır
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Özellik ve hedef
X = df.drop(['id', 'diagnosis'], axis=1)
y = df['diagnosis']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_enc = le.fit_transform(y)

# Model parametreleri
rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)

# Stratified K-Fold objesi
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Performans metriklerini tutmak için listeler
acc_scores, prec_scores, rec_scores, f1_scores = [], [], [], []

for train_index, test_index in skf.split(X, y_enc):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y_enc[train_index], y_enc[test_index]
    
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    
    acc_scores.append(accuracy_score(y_test, y_pred))
    prec_scores.append(precision_score(y_test, y_pred))
    rec_scores.append(recall_score(y_test, y_pred))
    f1_scores.append(f1_score(y_test, y_pred))

print(f"Accuracy: {np.mean(acc_scores):.4f} ± {np.std(acc_scores):.4f}")
print(f"Precision: {np.mean(prec_scores):.4f} ± {np.std(prec_scores):.4f}")
print(f"Recall: {np.mean(rec_scores):.4f} ± {np.std(rec_scores):.4f}")
print(f"F1-score: {np.mean(f1_scores):.4f} ± {np.std(f1_scores):.4f}")
```

. class_weight='balanced' nedir?
Modelin öğrenme aşamasında kullandığı bir parametredir.

Özellikle sınıflar arasında dengesizlik varsa (örneğin az sayıda pozitif, çok sayıda negatif örnek), modelin azınlık sınıfa yeterince önem vermesini sağlar.

Böylece model, azınlık sınıftaki örnekleri yanlış sınıflandırma cezasını artırır.

Bu sayede model, dengesiz veride daha adil ve dengeli öğrenir, azınlık sınıfı gözardı etmez.

Denetimli Öğrenme (Supervised Learning)
Ne var? Elinizde girdi (feature) ve ona karşılık gelen doğru çıktı (etiket, label) vardır.

Amaç: Modeli eğitip, yeni veriler için doğru çıktıyı tahmin etmek.

Örnekler:

Sınıflandırma (classification): E-posta spam mi değil mi? Kanser tanısı pozitif mi negatif mi?

Regresyon (regression): Ev fiyatı tahmini, sıcaklık tahmini gibi.

Veri: Etiketli veri gerektirir (her gözlem için doğru cevap var).

Denetimsiz Öğrenme (Unsupervised Learning)
Ne var? Elinizde sadece girdiler (özellikler) var, doğru çıktı/etiket yok.

Amaç: Verideki gizli yapıları, grupları, örüntüleri ortaya çıkarmak.

Örnekler:

Kümeleme (clustering): K-means, DBSCAN gibi algoritmalar. Müşterileri segmentlere ayırmak, anomali tespiti.

Boyut indirgeme (dimensionality reduction): PCA gibi.

Veri: Etiketsiz veriyle çalışır.

Özet:
Öğrenme Türü	Etiket Var mı?	Amaç	Örnekler
Denetimli	Evet	Yeni verinin etiketini tahmin	Random Forest, SVM
Denetimsiz	Hayır	Veri içindeki yapıyı keşfet	K-means, PCA

1. Stratified K-Fold Cross-Validation (CV) nedir?
Veri setini k parçaya (fold) böler, ama bu bölme sırasında her katmanda hedef sınıfların dağılımı (örneğin pozitif ve negatif sınıflar) orijinal veri setindeki gibi korunur (yani "stratified" yani katmanlı).

Modeli her katmanda eğitim ve test yaparak performansın tüm veri üzerinde tutarlı ve güvenilir bir şekilde ölçülmesini sağlar.

Ama modelin öğrenme aşamasına veya sınıf ağırlıklarına doğrudan bir etkisi yoktur. Sadece değerlendirme (validation) metodudur.

```python
from flask import Flask, request, jsonify
import joblib
import numpy as np
import pandas as pd

app = Flask(__name__)

# Model ve scaler yükle
model = joblib.load("cancer_model.pkl")
scaler = joblib.load("scaler.pkl")

@app.route('/predict', methods=['POST'])
def predict():
    # JSON formatında gelen veriyi al
    data = request.get_json()
    df = pd.DataFrame([data])  # Tek gözlem varsayımı

    # Gerekirse ölçekleme
    df_scaled = scaler.transform(df)

    # Tahmin
    prediction = model.predict(df_scaled)[0]
    prediction_label = "Malignant (M)" if prediction == 1 else "Benign (B)"

    return jsonify({
        "prediction": int(prediction),
        "diagnosis": prediction_label
    })

if __name__ == '__main__':
    app.run(debug=True)
```
