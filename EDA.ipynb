```python
# IQR yÃ¶ntemiyle aykÄ±rÄ± deÄŸerleri tespit edelim
Q1 = df['radius_mean'].quantile(0.25)
Q3 = df['radius_mean'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# AykÄ±rÄ± deÄŸerleri dÄ±ÅŸarÄ±da bÄ±rak
df = df[(df['radius_mean'] >= lower_bound) & (df['radius_mean'] <= upper_bound)]
```

```python
# AykÄ±rÄ± deÄŸerleri medyanla deÄŸiÅŸtir
median = df['radius_mean'].median()

df['radius_mean'] = df['radius_mean'].apply(
    lambda x: median if x < lower_bound or x > upper_bound else x
)
```

```python
df['radius_mean'] = df['radius_mean'].apply(
    lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x
)
```

```python
# M = 1, B = 0 gibi
df['diagnosis_encoded'] = df['diagnosis'].map({'M': 1, 'B': 0})

# Korelasyon matrisini tekrar oluÅŸtur
corr_matrix = df.corr(numeric_only=True)

# Åžimdi diagnosis_encoded ile olan korelasyonlarÄ± alabiliriz
target_corr = corr_matrix['diagnosis_encoded'].drop('diagnosis_encoded')
target_corr_sorted = target_corr.abs().sort_values(ascending=False)

print("Diagnosis ile en yÃ¼ksek korelasyona sahip deÄŸiÅŸkenler:")
print(target_corr_sorted.head(10))
```

```python
# Korelasyonu yÃ¼ksek olan Ã§iftleri bul
threshold = 0.9  # Ã§ok yÃ¼ksek korelasyon sÄ±nÄ±rÄ±
high_corr_pairs = []

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            colname1 = corr_matrix.columns[i]
            colname2 = corr_matrix.columns[j]
            corr_value = corr_matrix.iloc[i, j]
            high_corr_pairs.append((colname1, colname2, corr_value))

# SonuÃ§larÄ± yazdÄ±r
print("YÃ¼ksek korelasyonlu Ã¶zellik Ã§iftleri (>|0.9|):")
for col1, col2, corr_val in high_corr_pairs:
    print(f"{col1} - {col2}: {corr_val:.2f}")
```

```python
# Ã–rnek: Korelasyonu 0.95'ten bÃ¼yÃ¼k olanlardan birini sil
to_drop = set()
threshold = 0.95

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        corr_value = abs(corr_matrix.iloc[i, j])
        if corr_value > threshold:
            col1 = corr_matrix.columns[i]
            col2 = corr_matrix.columns[j]
            to_drop.add(col1)  # Ã¶rneÄŸin hep saÄŸdakini atÄ±yoruz

# Veri setinden Ã§Ä±kar
df_reduced = df.drop(columns=to_drop)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Ã–zellik ve hedef
X = df.drop(['id', 'diagnosis'], axis=1)
y = df['diagnosis']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_enc = le.fit_transform(y)

# Model parametreleri
rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)

# Stratified K-Fold objesi
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Performans metriklerini tutmak iÃ§in listeler
acc_scores, prec_scores, rec_scores, f1_scores = [], [], [], []

for train_index, test_index in skf.split(X, y_enc):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y_enc[train_index], y_enc[test_index]
    
    rf.fit(X_train, y_train)
    y_pred = rf.predict(X_test)
    
    acc_scores.append(accuracy_score(y_test, y_pred))
    prec_scores.append(precision_score(y_test, y_pred))
    rec_scores.append(recall_score(y_test, y_pred))
    f1_scores.append(f1_score(y_test, y_pred))

print(f"Accuracy: {np.mean(acc_scores):.4f} Â± {np.std(acc_scores):.4f}")
print(f"Precision: {np.mean(prec_scores):.4f} Â± {np.std(prec_scores):.4f}")
print(f"Recall: {np.mean(rec_scores):.4f} Â± {np.std(rec_scores):.4f}")
print(f"F1-score: {np.mean(f1_scores):.4f} Â± {np.std(f1_scores):.4f}")
```

. class_weight='balanced' nedir?
Modelin Ã¶ÄŸrenme aÅŸamasÄ±nda kullandÄ±ÄŸÄ± bir parametredir.

Ã–zellikle sÄ±nÄ±flar arasÄ±nda dengesizlik varsa (Ã¶rneÄŸin az sayÄ±da pozitif, Ã§ok sayÄ±da negatif Ã¶rnek), modelin azÄ±nlÄ±k sÄ±nÄ±fa yeterince Ã¶nem vermesini saÄŸlar.

BÃ¶ylece model, azÄ±nlÄ±k sÄ±nÄ±ftaki Ã¶rnekleri yanlÄ±ÅŸ sÄ±nÄ±flandÄ±rma cezasÄ±nÄ± artÄ±rÄ±r.

Bu sayede model, dengesiz veride daha adil ve dengeli Ã¶ÄŸrenir, azÄ±nlÄ±k sÄ±nÄ±fÄ± gÃ¶zardÄ± etmez.

Denetimli Ã–ÄŸrenme (Supervised Learning)
Ne var? Elinizde girdi (feature) ve ona karÅŸÄ±lÄ±k gelen doÄŸru Ã§Ä±ktÄ± (etiket, label) vardÄ±r.

AmaÃ§: Modeli eÄŸitip, yeni veriler iÃ§in doÄŸru Ã§Ä±ktÄ±yÄ± tahmin etmek.

Ã–rnekler:

SÄ±nÄ±flandÄ±rma (classification): E-posta spam mi deÄŸil mi? Kanser tanÄ±sÄ± pozitif mi negatif mi?

Regresyon (regression): Ev fiyatÄ± tahmini, sÄ±caklÄ±k tahmini gibi.

Veri: Etiketli veri gerektirir (her gÃ¶zlem iÃ§in doÄŸru cevap var).

Denetimsiz Ã–ÄŸrenme (Unsupervised Learning)
Ne var? Elinizde sadece girdiler (Ã¶zellikler) var, doÄŸru Ã§Ä±ktÄ±/etiket yok.

AmaÃ§: Verideki gizli yapÄ±larÄ±, gruplarÄ±, Ã¶rÃ¼ntÃ¼leri ortaya Ã§Ä±karmak.

Ã–rnekler:

KÃ¼meleme (clustering): K-means, DBSCAN gibi algoritmalar. MÃ¼ÅŸterileri segmentlere ayÄ±rmak, anomali tespiti.

Boyut indirgeme (dimensionality reduction): PCA gibi.

Veri: Etiketsiz veriyle Ã§alÄ±ÅŸÄ±r.

Ã–zet:
Ã–ÄŸrenme TÃ¼rÃ¼	Etiket Var mÄ±?	AmaÃ§	Ã–rnekler
Denetimli	Evet	Yeni verinin etiketini tahmin	Random Forest, SVM
Denetimsiz	HayÄ±r	Veri iÃ§indeki yapÄ±yÄ± keÅŸfet	K-means, PCA

1. Stratified K-Fold Cross-Validation (CV) nedir?
Veri setini k parÃ§aya (fold) bÃ¶ler, ama bu bÃ¶lme sÄ±rasÄ±nda her katmanda hedef sÄ±nÄ±flarÄ±n daÄŸÄ±lÄ±mÄ± (Ã¶rneÄŸin pozitif ve negatif sÄ±nÄ±flar) orijinal veri setindeki gibi korunur (yani "stratified" yani katmanlÄ±).

Modeli her katmanda eÄŸitim ve test yaparak performansÄ±n tÃ¼m veri Ã¼zerinde tutarlÄ± ve gÃ¼venilir bir ÅŸekilde Ã¶lÃ§Ã¼lmesini saÄŸlar.

Ama modelin Ã¶ÄŸrenme aÅŸamasÄ±na veya sÄ±nÄ±f aÄŸÄ±rlÄ±klarÄ±na doÄŸrudan bir etkisi yoktur. Sadece deÄŸerlendirme (validation) metodudur.
